```python
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image

import sys
import os
sys.path.append(os.path.abspath('..'))
from constants import BASE_DIR
```

```python
input_dir = BASE_DIR / "datasets/celebrities/mo_salah/cleaned"
image_paths = list(input_dir.glob("*.[jp][pn]*g"))
```

```python

def plot_images(image_paths, cols=5, figsize=(15, 10)):
    paths = list(image_paths)
    rows = -(-len(paths) // cols)  # Ceiling division trick
    plt.figure(figsize=figsize)
    for i, path in enumerate(paths, 1):
        plt.subplot(rows, cols, i)
        plt.imshow(Image.open(path))
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plot_images(image_paths, figsize=(15, 15))
```
